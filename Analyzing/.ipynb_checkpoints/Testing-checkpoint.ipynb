```python
def get_data_from_csv(data):
    V_peak = read_column_csv(data, 'PeakTrace')
    T_peak = read_column_csv(data, 'Time')
    PEAK = read_column_csv(data, 'Peak_Event')
    NOISE = read_column_csv(data, 'Noise_Event')
    return V_peak, T_peak, PEAK, NOISE

V_peak1, T_peak1, PEAK1, NOISE1 = get_data_from_csv(data1)
V_peak2, T_peak2, PEAK2, NOISE2 = get_data_from_csv(data2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for histograms
data1 = PEAK1
data2 = PEAK2

# Determine the common bin edges
bin_min = min(np.min(data1), np.min(data2))
bin_max = max(np.max(data1), np.max(data2))
num_bins = 20
bin_edges = np.linspace(bin_min, bin_max, num_bins + 1)

# Create a figure and set up subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 7))

# Plot the first histogram with log scale on y-axis
#axs[0].hist(data1, bins=bin_edges, alpha=0.7, color='blue', edgecolor='black')
axs[0].hist(data1, bins=int(max(data1)-min(data1))+1, range=(min(data1)-0.5, max(data1)+0.5), rwidth=0.8, edgecolor='black', color='red')

axs[0].set_title('Monopoles')
axs[0].set_yscale('symlog')

# Plot the second histogram with log scale on y-axis
#axs[1].hist(data2, bins=bin_edges, alpha=0.7, color='red', edgecolor='black')
axs[1].hist(data2, bins=int(max(data2)-min(data2))+1, range=(min(data2)-0.5, max(data2)+0.5), rwidth=0.8, edgecolor='black')    

axs[1].set_title('Neutrinos')
axs[1].set_yscale('symlog')
# Set the y-axis limit to be the same for both subplots
max_value = max(np.max(np.histogram(data1, bins=bin_edges)[0]), np.max(np.histogram(data2, bins=bin_edges)[0]))
for ax in axs:
    ax.set_ylim([0, max_value*1.2])

# Add labels and title to the figure
fig.suptitle(f'The Peak distribution for {num} particles at 1e{E}eV (Log Scale)')
plt.tight_layout()

# Show the plot
plt.show()
print(PEAK1)
print(PEAK2)
```

```python
dt = 0.2#ns
plt.figure(figsize=(15, 7))
noise1=np.mean(NOISE1)
plt.plot(times1*dt, abs(trace1)/noise1, alpha = 0.5, label = "Received signal")
plt.plot(amp_time1*dt, max_amp1/noise1, ".",alpha = 0.7, label = "Max signal amplitude of the shower")
plt.plot(T_peak1*dt, V_peak1/noise1, "x", color = "black",alpha=0.7, label = "Peak detection")

plt.plot(times1*dt, threshold+times1*0, ",", label = f"SNR = {threshold}")
plt.plot(times1*dt, 1+times1*0, ",", label = "SNR = 1")
plt.xlim(min(times1*dt), max(amp_time1*dt))
plt.ylim(5e-1, max( V_peak1/noise1)*1.5)
plt.legend()
plt.ylabel("SNR")
plt.xlabel("time (ns)")
plt.yscale("log")
#plt.ylim(1e-9, 1e-1)
#plt.xlim(100, 2100)
plt.show()
```
