```python
import numpy as np
import scipy.stats as stats
from scipy.optimize import curve_fit
def get_fit(x_data, y_data):
    x_data = np.log(x_data)
    y_data = np.log(y_data)
    # Define the model function
    def model_func(x, a, b):
        return a*x+b
    # Perform curve fitting
    popt, pcov = curve_fit(model_func, x_data, y_data)
    # Calculate residuals
    y_fit = model_func(x_data, *popt)
    residuals = y_data - y_fit
    # Calculate the reduced chi-square value
    print(f"Best-fit parameters: {popt}")
    return popt
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Generate a synthetic waveform with noise
time = np.linspace(0, 1, 1000)
signal_freq = 50  # Frequency of the signal component
noise_amplitude = 0.2
signal_amplitude = 1.0
waveform = signal_amplitude * np.sin(2 * np.pi * signal_freq * time) + noise_amplitude * np.random.randn(len(time))

# Compute the power spectral density (PSD) of the waveform
frequencies, psd = signal.welch(waveform, fs=1.0, nperseg=256)

# Plot the waveform and its power spectral density
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))
ax1.plot(time, waveform)
ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')
ax1.set_title('Waveform')

ax2.plot(frequencies, psd)
ax2.set_xlabel('Frequency')
ax2.set_ylabel('Power Spectral Density')
ax2.set_title('Power Spectral Density')

plt.tight_layout()
plt.show()

# Extract the noise component from the waveform
noise_index = np.argmax(psd)  # Index of the peak in the PSD represents the noise
noise = np.random.normal(loc=0, scale=np.sqrt(psd[noise_index]), size=len(time))

# Remove the noise from the waveform
clean_waveform = waveform - noise

# Plot the clean waveform
plt.figure()
plt.plot(time, clean_waveform)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Clean Waveform')
plt.show()
```
